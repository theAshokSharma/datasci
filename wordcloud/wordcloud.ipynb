```python
# Wordcloud

from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd
from PIL import Image, ImageFilter
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
# Reads text file
with open(r"romeo_and_juliet.txt", "r") as file:
   text = file.read()
```

```python
# read the image file and create a mask_img

# Each pixel in a color image is represented digitally by three numbers corresponding to the red, green, and blue 
# values of that pixel. Thresholding is the process of converting all the pixels to either the maximum or minimum value 
# depending on whether they’re higher or lower than a certain number. It’s easier to do this on a grayscale image:

# You achieve thresholding by calling .point() to convert each pixel in the grayscale image into either 255 or 0.
# you can learn more about image processing at https://realpython.com/image-processing-with-the-python-pillow-library/

filename = "rom_jul.png"
with Image.open(filename) as img:
    img.load()

# comment this line for the before submitting
img.show()

img_gray = img.convert("L")

threshold = 125
img_threshold = img_gray.point(
    lambda x: 255 if x > threshold else 0
)
# img_threshold.show()
mask_img = np.array(img_threshold)
```

```python
comment_words = ''
stopwords = set(STOPWORDS)
```

```python
wordcloud = WordCloud(width = 3000,
                      height = 3000,
                      background_color ='black',
                      stopwords = stopwords,
                      max_words=200,
                      mask=mask_img,
                      min_font_size = 10).generate(text)
```

```python
plt.imshow(wordcloud,
           interpolation='bilinear')
plt.axis("off")
```
